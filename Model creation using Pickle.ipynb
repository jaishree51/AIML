```python
from sklearn.tree import DecisionTreeClassifier

# Define input and output data
input_dataset = [[25, 1], [30, 0], [45, 1], [35, 0]]  # Example features: [age, likes_action_movies]
output_dataset = ['Yes', 'No', 'Yes', 'No']          # Example labels: interested or not

# Train the model
movie_model = DecisionTreeClassifier()
movie_model.fit(input_dataset, output_dataset)

# Make predictions
movie_interest = movie_model.predict([[9, 1], [33, 0]])
print(movie_interest)

```
